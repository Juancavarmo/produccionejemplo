```r
# Multiplot
multiplot <- function(..., plotlist=NULL, file, cols=1, layout=NULL) {
  library(grid)
  
  # Make a list from the ... arguments and plotlist
  plots <- c(list(...), plotlist)
  
  numPlots = length(plots)
  
  # If layout is NULL, then use 'cols' to determine layout
  if (is.null(layout)) {
    # Make the panel
    # ncol: Number of columns of plots
    # nrow: Number of rows needed, calculated from # of cols
    layout <- matrix(seq(1, cols * ceiling(numPlots/cols)),
                     ncol = cols, nrow = ceiling(numPlots/cols))
  }
  
  if (numPlots==1) {
    print(plots[[1]])
    
  } else {
    # Set up the page
    grid.newpage()
    pushViewport(viewport(layout = grid.layout(nrow(layout), ncol(layout))))
    
    # Make each plot, in the correct location
    for (i in 1:numPlots) {
      # Get the i,j matrix positions of the regions that contain this subplot
      matchidx <- as.data.frame(which(layout == i, arr.ind = TRUE))
      
      print(plots[[i]], vp = viewport(layout.pos.row = matchidx$row,
                                      layout.pos.col = matchidx$col))
    }
  }
}
```

```r
g1<-ggplot(data = data_oil, aes(x=Country, y = Prod.Oil)) + geom_boxplot()
g2<-ggplot(data = data_oil, aes(x = Prod.Oil)) + geom_histogram(bins = 25, fill = 'white', col = 'darkgreen')
g3<-ggplot(data=data_oil, aes(sample=Prod.Oil)) + stat_qq() + stat_qq_line(col = 'red', size = 1)
layout <- matrix(c(1, 1, 1, 2, 2, 3), nrow = 2, byrow = TRUE)
multiplot(g1, g2, g3, layout = layout)
```

```r
### Un Gráfico adicional
paises2<-paises[,2:5]
names(paises2)<-c('Lon','Lat','N_Country','Country')
geo_data<-merge(data_oil,paises2,by='Country')
geo_data_P<-aggregate(Prod.Oil~Lon+Lat+Country, data = geo_data, FUN = sum)
```

```r
library(ggrepel)
```

```r
ggplot(data = geo_data_P, aes(x = Lon, y = Lat, col = Country)) + 
  geom_point(aes(size = Prod.Oil)) +
  geom_text_repel(aes(x = Lon, y = Lat,
                      label = paste0(Country,',',round(Prod.Oil*1e-6,2))))
```

```r
Point_Box<- matrix(c(min(geo_data_P$Lon)-10,min(geo_data_P$Lat)-10,
           max(geo_data_P$Lon)+10,max(geo_data_P$Lat)+10), nrow = 2)
colnames(Point_Box)<-c('min','max')
```

```r
library(ggmap)
```

```r
WGG <- get_map(c(Point_Box), maptype="roadmap", source="google")
```

```r
ggmap(WGG)+ geom_point(data = geo_data_P, aes(x = Lon, y = Lat, size = Prod.Oil,
                       col = Country)) +
  geom_text_repel(data = geo_data_P, aes(x = Lon, y = Lat,
                      label = paste0(Country,',',round(Prod.Oil*1e-6,2))))
```
